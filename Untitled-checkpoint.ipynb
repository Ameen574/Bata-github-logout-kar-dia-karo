```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text,"lxml")
soup
```

```python
title= soup.find_all("a",class_="title")
title
```

```python
for item in title:
    print(item.text.strip())
```

```python
for item in title:
    title_name = []

    title_name.append(item.text.strip())
```

```python
title_name
```

```python
price = soup.find_all("h4",class_="price float-end card-title pull-right")
price
```

```python
for item in price :
    print(item.text.strip())
```

```python
pro_price = []
for item in price :
    pro_price.append(item.text.strip())
```

```python
pro_price
```
